```python
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
sns.set_style('dark')
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc)
```

```python
#note harus dijadikan data matrix WAJIB!!!
fp = flights.pivot_table(index='month',columns='year',values='passengers')
sns.heatmap(fp,cmap='magma')
```

```python
sns.clustermap(fp,standard_scale=1) #melakukan standarisasi untuk legend
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
```

```python
iris['species'].unique()
```

```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
g = sns.FacetGrid(tips, col="day",  row="smoker")
g = g.map(plt.hist, "total_bill")
```

```python
#menggunakan 2 argument

g = sns.FacetGrid(tips, col="time",  row="smoker",hue='sex')
# Notice hwo the arguments come after plt.scatter call
g = g.map(plt.scatter, "total_bill", "tip").add_legend()
```

# LM plot


```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',markers=['o','v'])
```

```python
sns.lmplot(x="total_bill", y="tip", hue='sex', col="day",data=tips,palette='seismic',aspect=0.6,size=8)
```
