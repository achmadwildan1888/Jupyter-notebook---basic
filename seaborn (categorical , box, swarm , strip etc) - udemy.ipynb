```python
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
# Safe to ignore warnings
```

```python
sns.jointplot(x='tip', y='total_bill',data=tips,kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill',data=tips,kind='kde') #kde menunjukan frekuensi kapan titik itu bertemu
```

```python
sns.pairplot(tips,hue='smoker') #hue diisi dengan categorical
```

# Categorical plot 

```python
import numpy as np
sns.barplot(data=tips,x='sex', y='total_bill',estimator=np.std) #x categorical, y = numerical
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True) #disatukan
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)  #jitter agar datanya tidak jadi strip only dan tau distribusinya 
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex', split=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex', split=True) #kombinasi dr atas
```

```python
#or just use factor plot to use up to you

sns.factorplot(x='day',y='total_bill',data=tips,hue='sex', kind='swarm')
```
